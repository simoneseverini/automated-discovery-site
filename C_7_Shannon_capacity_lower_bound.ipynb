```python
# EVOLVE-BLOCK-START
import random
import time
from itertools import product
import array

def find_independent_set(n: int, time_limit: float, current_best_solution: list[tuple]) -> list[tuple]:
    start_t, num_v, p7 = time.time(), 7**n, [7**k for k in range(n)]
    adj = []
    for i in range(num_v):
        nb = [i]
        for k in range(n):
            vk = (i // p7[k]) % 7
            s1, s2 = ((vk - 1) % 7 - vk) * p7[k], ((vk + 1) % 7 - vk) * p7[k]
            nb += [r + s1 for r in nb] + [r + s2 for r in nb]
        adj.append(array.array('H' if num_v < 65536 else 'I', nb))

    forbidden_count = array.array('H', [0] * num_v)
    free_vertices = list(range(num_v))
    pos_in_free = array.array('i', range(num_v))
    current_S = []

    def add_v(idx):
        current_S.append(idx)
        for nb in adj[idx]:
            if forbidden_count[nb] == 0:
                p = pos_in_free[nb]
                v_last = free_vertices.pop()
                if nb != v_last:
                    free_vertices[p], pos_in_free[v_last] = v_last, p
            forbidden_count[nb] += 1

    def remove_v(idx):
        for nb in adj[idx]:
            forbidden_count[nb] -= 1
            if forbidden_count[nb] == 0:
                pos_in_free[nb] = len(free_vertices)
                free_vertices.append(nb)

    seed = [sum(v[k]*p7[k] for k in range(n)) for v in current_best_solution] if current_best_solution else \
           [sum(v[k]*p7[k] for k in range(n)) for v in product((0,2,4), repeat=n)]
    for idx in seed:
        if forbidden_count[idx] == 0: add_v(idx)
    while free_vertices: add_v(random.choice(free_vertices))

    best_S_idx = current_S[:]
    stall_count, max_stall = 0, int(len(best_S_idx) * 20)

    while time.time() - start_t < time_limit - 0.4:
        # Optimized Local Search: 1-swaps on plateau
        if not current_S: break

        # 1. Pick random vertex u in S
        idx_in_S = random.randrange(len(current_S))
        u = current_S[idx_in_S]

        # 2. Criticality check: does removing u free any *other* vertex?
        # A vertex nb is freed if forbidden_count[nb] == 1 (it is blocked only by u).
        # We check neighbors; if nb == u, it is freed but that's trivial. We want nb != u.
        is_critical = False
        for nb in adj[u]:
            if nb != u and forbidden_count[nb] == 1:
                is_critical = True
                break

        # 3. Move logic
        if is_critical:
            # Perform swap/improve
            current_S[idx_in_S] = current_S[-1]; current_S.pop()
            remove_v(u)

            # Fill greedily from free vertices (includes u and newly freed neighbors)
            # This effectively performs a (1, k)-swap.
            while free_vertices:
                v = random.choice(free_vertices)
                add_v(v)

            if len(current_S) > len(best_S_idx):
                best_S_idx = current_S[:]
                max_stall = int(len(best_S_idx) * 20)
                stall_count = 0
            else:
                stall_count += 1
        else:
            # u is stable (removing it frees nothing new).
            # Skipping removal saves time, but we count it as a stall.
            stall_count += 1

        # 4. Perturbation if stuck
        if stall_count > max_stall:
            # Ruin phase: remove ~10%
            num_ruin = max(1, int(len(current_S) * 0.1))
            for _ in range(num_ruin):
                if not current_S: break
                ridx = random.randrange(len(current_S))
                rem_v = current_S[ridx]; current_S[ridx] = current_S[-1]; current_S.pop()
                remove_v(rem_v)

            # Recreate
            while free_vertices:
                add_v(random.choice(free_vertices))

            if len(current_S) > len(best_S_idx):
                best_S_idx = current_S[:]
                max_stall = int(len(best_S_idx) * 20)
            stall_count = 0

    return [tuple((idx // p7[k]) % 7 for k in range(n)) for idx in best_S_idx]
# EVOLVE-BLOCK-END
import math
import matplotlib.pyplot as plt
import io
import base64

def serialize(fig: plt.Figure) -> str:
    buf = io.BytesIO()
    fig.savefig(buf, format='png', dpi=100)
    buf.seek(0)
    img = base64.b64encode(buf.getvalue()).decode('utf-8')
    buf.close()
    plt.close(fig)
    return img

def verify_independent_set(solution: list[tuple], n: int) -> tuple[bool, str]:
    """
    Verifies that the solution is a valid independent set in C_7^n.
    Conditions:
    1. All elements are tuples of length n.
    2. All elements are in Z_7^n.
    3. Elements are unique.
    4. NO two elements are adjacent.
       u ~ v iff for all i, u_i - v_i in {0, 1, -1} mod 7.
    """
    if not isinstance(solution, list):
        return False, "Solution is not a list."

    seen = set()

    for idx, pt in enumerate(solution):
        if not isinstance(pt, tuple) or len(pt) != n:
            return False, f"Item {idx} is not a tuple of length {n}."
        for x in pt:
            if not isinstance(x, int) or not (0 <= x < 7):
                return False, f"Item {idx} contains invalid coordinate {x} (must be 0..6)."

        if pt in seen:
            return False, f"Duplicate item found: {pt}"
        seen.add(pt)

    # Adjacency Check (O(S^2 * n))
    # Two vertices collide if their difference vector only contains {0, 1, 6} (mod 7).
    # i.e., diff is NOT in {2, 3, 4, 5} for any coordinate.

    # Optimization: Convert to list of lists for indexing speed
    sol_list = list(solution)
    num_pts = len(sol_list)

    for i in range(num_pts):
        u = sol_list[i]
        for j in range(i + 1, num_pts):
            v = sol_list[j]

            # Check if u ~ v
            is_adjacent = True
            for k in range(n):
                diff = (u[k] - v[k]) % 7
                if diff not in (0, 1, 6):
                    # Found a coordinate where they differ significantly -> NOT adjacent
                    is_adjacent = False
                    break

            if is_adjacent:
                return False, f"Adjacency violation between {u} and {v}."

    return True, "Valid"

def visualize(solution: list[tuple], n: int) -> str:
    """
    Visualizes the solution by projecting onto the first two dimensions.
    """
    if not solution:
        return ""

    # Count occurrences in the 2D plane (x0, x1)
    grid = [[0]*7 for _ in range(7)]
    for pt in solution:
        grid[pt[0]][pt[1]] += 1

    fig, ax = plt.subplots(figsize=(5, 5))
    im = ax.imshow(grid, cmap='viridis', origin='lower', vmin=0)

    # Annotate counts
    for x in range(7):
        for y in range(7):
            count = grid[x][y]
            if count > 0:
                ax.text(y, x, str(count), ha='center', va='center', color='white' if count > max(1, len(solution)//100) else 'black')

    ax.set_xticks(range(7))
    ax.set_yticks(range(7))
    ax.set_xlabel("Coord 1")
    ax.set_ylabel("Coord 0")
    ax.set_title(f"Projection of IS (Size {len(solution)}) onto first 2 dims")
    fig.colorbar(im)

    return serialize(fig)

def evaluate(data: dict) -> tuple[dict, dict]:
    # Set problem dimension.
    # n=5 is chosen for the evolutionary loop as it is fast (16k vertices).
    # Logic evolved for n=5 typically generalizes to n=6.
    n = 5
    time_limit = 20.0 # Strict time limit for evolution

    # Retrieve parent's best solution to seed the search
    parent_sol = data.get('parent_outputs', {}).get('best_solution', [])

    # Run the evolvable code
    try:
        proposed_solution = find_independent_set(n, time_limit, parent_sol)
    except Exception as e:
        return {'score': -10000}, {'error': str(e)}

    # Verify
    is_valid, message = verify_independent_set(proposed_solution, n)

    if not is_valid:
        score = -10000
        print(f"Invalid solution: {message}")
    else:
        score = len(proposed_solution)

    # Visualize
    image_data = visualize(proposed_solution if is_valid else [], n)

    # Calculate Shannon capacity lower bound
    theta = score ** (1/n) if score > 0 else 0

    return (
        {'size': float(score), 'theta_lb': theta},
        {'best_solution': proposed_solution if is_valid else [], 'image': image_data}
    )
```
